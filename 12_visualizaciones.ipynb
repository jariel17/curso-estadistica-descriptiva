```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
# scatterplot
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
# joint plot
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

```python
# joint plot
sns.jointplot(data=iris, x='petal_width', y='petal_length', hue='species')
```

## Reto crear otros graficos

https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
# box plot

sns.boxplot(x = 'species', y = 'sepal_length', data=iris)
```

https://seaborn.pydata.org/generated/seaborn.barplot.html

```python
iris_medians = iris.groupby('species').median()
iris_medians
```

```python
# barplot


sns.barplot(data = iris_medians, x ='petal_width', y='petal_length') 
```
